```python
# import package
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
# import data
df=pd.read_csv('data/AirPassengers.csv')
df.head()
```

```python
data=df[['#Passengers']].values
data
```

```python
# data preparation
```

```python
sc=MinMaxScaler(feature_range=(0,1))
data_scaled=sc.fit_transform(data)
```

```python
data_scaled
```

```python
X_data=[]
y_data=[]
look_back=1
for i in range(len(data_scaled)-1):
    X_data.append(data_scaled[i:(i+look_back),0])
    y_data.append(data_scaled[i+look_back,0])
X_data, y_data=np.array(X_data),np.array(y_data)
print(X_data)
print(y_data)
```

```python
# reshaping
X_data=np.reshape(X_data,(X_data.shape[0],X_data.shape[1],1))
```

```python
# split data
train_size=int(len(X_data)*0.7)
test_size=len(X_data)-train_size
X_train=X_data[0:train_size,:]
X_test=X_data[train_size:len(X_data),:]
y_train=y_data[0:train_size]
y_test=y_data[train_size:len(y_data)]
```

```python
# build RNN
model=Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))

model.add(LSTM(units=50, return_sequences=True))
model.add(LSTM(units=50))

model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

```python
# fit model
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
# analyze results
y_pred=model.predict(X_test)
```

```python
# visualize results
plt.plot(y_test, color='red', label='nombre de passagers rel')
plt.plot(y_pred, color='blue', label='nombre de passagers prvus')
plt.title('Les passagers prvus vs rel')
plt.xlabel('temps')
plt.ylabel('passagers')
plt.legend()
```
